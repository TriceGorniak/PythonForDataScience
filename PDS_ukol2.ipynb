VYPRACOVÁNÍ ÚKOLU

```python
##nahrání potřebných knihoven
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split ##metoda pro rozdělení dat na testovací a trénovací sadu
from sklearn import datasets
from sklearn.svm import LinearSVC, SVC
from sklearn.metrics import (accuracy_score,confusion_matrix,ConfusionMatrixDisplay)
```

```python
##část kodu ze zadání - nahrání datasetu přímo z modulu
digits = datasets.load_digits()
y = digits.target
data = digits.images

##část kódu ze zadání - převod dat na flat
n_samples = len(data)
data = data.reshape([n_samples, -1])
```

1) Rozděl data na testovací a trénovací pomocí metody train_test_split, přičemž nastav parametry test_size=0.3 a random_state=42. Použij stejné názvy proměnných, jaké jsme používali v lekci, tj. X_train, X_test, y_train, y_test.

```python
##rozdělení na testovací a trénovací sadu 
X = data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

2) Použij klasifikátor SVC a jako kernel použij linear. Parametr decision_function_shape nenastavuj a využij výchozí hodnotu, tj. ovo. Nenastavuj ani žádné další parametry.

3) Natrénuj model s využitím metody fit() 

4) a nakonec metodou predict() vytvoř pole predikcí pro testovací data, které pojmenuj y_pred.

```python
clf = SVC(kernel="linear")##SVC linear 
clf.fit(X_train, y_train)##trénink modelu metodou fit()
y_pred = clf.predict(X_test) ##metoda predict()
```

5) Zjisti hodnotu metriky accuracy (měla by ti vyjít přibližně 0.9796) 

```python
accuracy_score(y_test, y_pred)
```

6) a matici záměn.

```python
ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test)
```
